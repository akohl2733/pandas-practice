```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",lw=1,ls="-",marker="o",markersize=10,
         markerfacecolor="Yellow",markeredgewidth=3,markeredgecolor="Green")
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",lw=1,ls="--")

axes.set_xlim([0,1])
axes.set_ylim([0,1])
```

```python
plt.scatter(x,y)
```
